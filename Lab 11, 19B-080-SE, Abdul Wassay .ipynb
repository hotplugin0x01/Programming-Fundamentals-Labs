```python
class ResidentialHouses:
    def __init__(self,location,rooms,parking,price,type):
        self.location=location
        self.rooms =rooms
        self.parking=parking
        self.price=price
        self.type=type
    def details(self):
        print('The price of {} type house located in {} with {} rooms and parking {} is {}.'.format(self.type,
                                                                                                    self.location,self.rooms,
                                                                                                   self.parking, self.price))
    def facility(self,facility):
        print('This house have facility of {}.'.format(facility))
        
class Bachelors(ResidentialHouses):
    pass

class Family(ResidentialHouses):
    pass

class Flates(ResidentialHouses):
    pass

class Apartments(ResidentialHouses):
    pass
```

```python
house=Bachelors("Johar",3,"not available",1000000,"Bachelors")
house.details()
house.facility('Garden')
```

```python
house=Family("DHA",5,"available",2500000,"Family")
house.details()
```

```python
house=Flates("Bahria Town",3,"availabe",1500000,"single flates")
house.details()
```

```python
house=Apartments("Defence",4,"available",3000000,"Apartment")
house.details()
house.facility('swimming pool')
```

```python
class Plant:
    def __init__(self,name,type,color,area,taste):
        self.name=name
        self.type=type
        self.color=color
        self.area=area
        self.taste=taste
    def details(self):
        print('The name of plant is:',self.name)
        print('This plant is {} and the colour of plant is {}.'.format(self.type,self.color))
        print('This plant tastes {} and it is {}.'.format(self.taste,self.area))
                
class Strawberry(Plant):
    pass

class Watermelon(Plant):
    pass

class Grapes(Plant):
    pass

class Rhubarb(Plant):
    pass

class Hydrangea(Plant):
    pass
```

```python
plant1=Strawberry("Strawberry","Fruit","Red","Local","Sweet")
plant1.details()
```

```python
plant2=Watermelon("Watermelon","Fruit","Red","Local","Sweet")
plant2.details()
```

```python
plant3=Grapes("Grapes","Fruit","Black","Local","Sweet")
plant3.details()
```

```python
plant4=Rhubarb("Rhubarb","Poisonous","Pink","imported","Sour")
plant4.details()
```

```python
plant5=Hydrangea("Hydrangea","Poisonous","Purple","imported","Sour")
plant5.details()
```

```python
class Student:
    def __init__(self,roll_num,student_name,department,section,address,contact):
        self.roll_num=roll_num
        self.student_name=student_name
        self.department=department
        self.section=section
        self.address=address
        self.contact=contact
    def details(self):
        print("The roll no. of student is:",self.roll_num)
        print("The name of student is:",self.student_name)
        print("The belongs to",self.department,'department.')
        print("The section of department is:",self.section)
        print("The address of student is:",self.address)
        print("Contact number of student is:",self.contact)

class Teacher(object):
    def __init__(self,teacher_id,teacher_name,department,rank,salary,address,number):
        self.teacher_id=teacher_id
        self.teacher_name=teacher_name
        self.department=department
        self.rank=rank
        self.salary=salary
        self.address=address
        self.number=number
    def details(self):
        print("The ID of teacher is:",self.teacher_id)
        print("Teacher's name is:",self.teacher_name)
        print("Teacher demonstrates in",self.department,'department.')
        print("The rank of teacher is",self.rank)
        print("Salary of teacher is:",self.salary)
        print("The address of teacher is:",self.address)
        print("Contact number is:",self.number)
```
